```python
health_df['TotalMinutes']=health_df['VeryActiveMinutes']+health_df['FairlyActiveMinutes']+health_df['LightlyActiveMinutes']+health_df['SedentaryMinutes']
health_df
```

I drew the following conclusions from data:

1. The Population(users) spend around approx 80% of their time sedentary(Inactive) while wearing the fitbit bands

2. On a average the users only take around 7637 steps which is borderline steps for being "somewhat active", if we take closer look around 25 Percentile of people lives extremely inactive life.

```python
from matplotlib import pyplot as plt
from matplotlib import  style 
style.use('ggplot')
plt.figure(figsize=(12,10))
plt.scatter(health_df['TotalMinutes'],health_df['SedentaryMinutes'],c=health_df['SedentaryMinutes'])
mean_SedentaryMinutes=991.210638
plt.axhline(mean_SedentaryMinutes,color='black',label='mean sedentaryminutes')
plt.title("Amount of time users do nothing")
plt.xlabel("Total Minutes")
plt.ylabel("Sedentary Minutes")
plt.grid(True,color="White")
plt.colorbar()
plt.legend()
plt.plot()
```

This is a positive correlation graph

As the amount of total minutes increase, the Sedentary minutes increase too, this shows that users proscatinate a lot or do nothing despite wearing fitbits which is a discouraging fact considering the fact Fitbits are made to make someone lifestyle healthier

# The amount of Calories burnt for Total amount of steps taken

```python
from matplotlib import pyplot as plt 
from matplotlib import style 
import numpy as np 
style.use("ggplot")
plt.figure(figsize=(12,10))
plt.scatter(health_df["TotalSteps"],health_df['Calories'],c=health_df['Calories'])
plt.title("The amount of Calories burnt for Total amount of steps taken")
plt.xlabel("Total Amount Of Steps")
plt.ylabel("Calories")
mean_calories=2303.609574
plt.axhline(mean_calories, color = "Black", label = "Mean Calories Burnt")
plt.grid(True,color="Black")
plt.colorbar()
plt.legend()
z = np.polyfit(health_df['TotalSteps'], health_df['Calories'], 1)
p = np.poly1d(z)
plt.plot(health_df['TotalSteps'],p(health_df['TotalSteps']),"r--")
```


note down some interesting observations:


There is an outlier above the right hand side of mean Calories Burnt line, it could occur due to human or instrument(Fitbit)
error

It's postive correlation graph

On an average, the users are burning 2303 calories, but the result heavily varies if you look at 15,000 steps, 
the blue cluster is burning approx 2000 calories, dark and light greenish cluster are buring 3000-4000 calories, 
some yellow spots are burning >4000 calories, it's due to the diffrence between VeryActive, FairyActive and LightlyActive users
